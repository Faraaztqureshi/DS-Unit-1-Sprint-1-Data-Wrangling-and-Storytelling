### Convert `int_rate`

Define a function to remove percent signs from strings and convert to floats

```python
## our data has % sign and leading spaces
df['int_rate'][:10]
```

```python
## if we look we have space in front and percent sign
df['int_rate'][:10].iloc[0]
```

```python
## we can use .strip and pass in what we want stripped
## default is the white space
int_rate = ' 14.47%'
```

```python
## we can chain strip 
## the first strip removes white space from beginning or end
## the second strip specifically removes % signs from the beginning or end
int_rate.strip().strip('%')
```

```python
## the aboce is still a string (see quotes)
type(int_rate.strip().strip('%'))
```

```python
## this function does what we want now to apply it
def int_rate_to_float(int_rate_string):
  return float(int_rate_string.strip().strip('%'))
```

Apply the function to the `int_rate` column

```python
## so we grab the int rate column and use .apply(name of function)
df1['int_rate'] = df1['int_rate'].apply(int_rate_to_float)
```

```python
## check out what we did now our object is a float
df1.head()
df1.dtypes
```

Clean the title and handle missing values

```python
## lets start to figure out what to do here by starting small
examples = ['owner', 'Supervisor', '  Project Manager', np.NaN]
```

```python
## title casing caps first letter and undercases all others
## strip removes leading trailing white space
def clean_title(title):
  if isinstance(title,str):
    return title.strip().title()
  else:
    return "Unknown"

[clean_title(item) for item in examples]
```

```python
## now that our function works we can resave our data after applying the title cleaning
df1['emp_title'] = df1['emp_title'].apply(clean_title)
```

```python
df1['emp_title'].value_counts()
```

### Create `emp_title_manager`

pandas documentation: [`str.contains`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.str.contains.html)

```python
## dataframe filtering
## we want to split our dataframe based on specific conditions (subsetting)
## lets make a new column
df1['emp_title_manager'] = df1['emp_title'].str.contains("Manager")
```

```python
## the above code will return True if there is "Manager" in the employment title
```

```python
## lets make a condition
## this condition will retain true rows and omit false
condition = (df1['emp_title_manager'] == True)
condition
```

```python
## if we apply this to our dataframe we will get back only those with manager in their title
managers = df1[condition]
managers.head()
```

```python
## we can do the same for non managers
## the ~ means opposite of
non_managers = df1[~condition]
non_managers.head()
```

```python
## now we can use these subsets
## we can get basic stats
managers.describe()
```
